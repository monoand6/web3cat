```python
addresses = [
    "0x2FAF487A4414Fe77e2327F0bf4AE2a264a776AD2",
]
tokens = [
    "WETH", "ETH", "USDC", "USDT", "DAI", "WBTC", 
    "LINK",  "AAVE",  "UNI", "1INCH", 
    "EURT", "SNX", "MATIC", "BUSD", "KNC", "CRO"]

base_tokens = ["USDC"]

v = View().portfolio_by_token(
    addresses = addresses,
    tokens = tokens,
    base_token = "USDC",
    start=datetime(2022, 11, 1),
    end=datetime(2022, 11, 11),
    numpoints=50
)
v.show()
```
